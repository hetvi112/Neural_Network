```python
x=df1.drop(columns='label',axis=1)
y=df1['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x_tarin,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_tarin.shape,x_test.shape)
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python

model=keras.Sequential([
  keras.layers.Flatten(input_shape=(30,)),
  keras.layers.Dense(20,activation='relu'),
  keras.layers.Dense(2,activation='sigmoid') # Changed output layer to 1 unit
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
histeory=model.fit(x_tarin,y_train,validation_split=0.1,epochs=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_tarin_std=scaler.fit_transform(x_tarin)
```

```python
x_test_std=scaler.transform(x_test)
```

```python
print(x_tarin_std)
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
histeory=model.fit(x_tarin_std,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(histeory.history['accuracy'])
plt.plot(histeory.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','val'],loc='lower right')
plt.show()
```

```python
plt.plot(histeory.history['loss'])
plt.plot(histeory.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
```

```python
loss,accuracy=model.evaluate(x_test_std,y_test)
```
